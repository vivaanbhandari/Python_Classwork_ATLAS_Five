```python
# Setup
import pandas as pd
```

```python
# Function to accept a list of reports 
# Returns a mega report dataframe (of all divisions)

def get_report_df(report_list): 
    
    # Create an empty mega report dataframe
    mega_report_df = pd.DataFrame()

    # For Loop
    # For Each Division Dictionary in the Report: 
    #########################################################
    
    for div_dictionary in report_list: 
    
        # Create a Student Scores Dataframe 
        scores_df = pd.DataFrame(div_dictionary["scores"])
        
        # Create an Emails Dataframe
        emails_df = pd.DataFrame(div_dictionary["emails"])
        
        # Merge the two dataframes together
        merged_df = pd.merge(scores_df, emails_df, how = "left", on = ["student_name"])  
        
        # Add a class column
        merged_df["class"] = div_dictionary["class"]
        
        # Add a division column
        merged_df["division"] = div_dictionary["division"]
        
        # Add a teacher column
        merged_df["teacher"] = div_dictionary["teacher"]
        
        # Add this merged dataframe to Mega Report Dataframe
        mega_report_df = pd.concat([mega_report_df, merged_df], ignore_index= True )
        
    #########################################################
    
    # Add the Total Score, Average Score, Pass/Fail, Character Grade Column
    mega_report_df["total_score"] = mega_report_df["physics_scores"] + mega_report_df["chemistry_scores"] + mega_report_df["information_technology_scores"] + mega_report_df["math_scores"]
    mega_report_df["average_score"] = mega_report_df["total_score"]/4

    
    mega_report_df.loc[mega_report_df["average_score"]>=35, "status"] = "pass"
    mega_report_df.loc[mega_report_df["average_score"]<35, "status"] = "fail"
    
    mega_report_df.loc[mega_report_df["average_score"] >= 90, "character_grade"] = "A"
    mega_report_df.loc[(mega_report_df["average_score"] >= 70) &(mega_report_df["average_score"] <90), "character_grade"] = "B"
    mega_report_df.loc[(mega_report_df["average_score"] >= 50) &(mega_report_df["average_score"] <70), "character_grade"] = "C"
    mega_report_df.loc[(mega_report_df["average_score"] >= 35) &(mega_report_df["average_score"] <50), "character_grade"] = "D"
    mega_report_df.loc[mega_report_df["average_score"] < 35, "character_grade"] = "F"
    
    # Return the Mega Report DataFrame
    return mega_report_df
```

```python
main_report = get_report_df(report_database)
main_report
```

# We want to know how many students passed and how many students failed in each division? 

```python
pass_fail_df = main_report.groupby(["division", "status"]).agg(total_students = ("status", "count"))
pass_fail_df
```

# We want to know what is the mean total score by division?

```python
mean_total_score_df = main_report.groupby(["division"]).agg(mean_total_score = ("total_score", "mean"))
mean_total_score_df
```
